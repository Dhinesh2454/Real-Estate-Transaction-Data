```python
### fill nul

mean_columns=['QS_ROOMS','QS_BATHROOM','QS_BEDROOM']
target_colum='QS_OVERALL'
def fill_na(rows):
    if pd.isnull(rows[target_colum]):
        return rows[mean_columns].mean()
    else:
        return rows[target_colum]
        
```

```python
df[target_colum]=df.apply(fill_na,axis=1)
```
